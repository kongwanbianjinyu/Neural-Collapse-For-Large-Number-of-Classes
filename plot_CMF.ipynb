```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

# Train ACC

```python
pd.read_csv("CMF_train.csv")
```

```python
df = pd.read_csv("CMF_train.csv")
```

```python
print(df.keys())
```

```python
resnet18_learnable_train = df['resnet18-cifar100-d10-tau10.0-epoch200-lr0.1 - acc/train']
CMF_resnet18_m1_train = df["CMF-resnet18-cifar100-d10-tau10.0-epoch200-lr0.1-momentum-0.1 - acc/train"]
CMF_resnet18_m9_train = df["CMF-resnet18-cifar100-d10-tau10.0-epoch200-lr0.1-momentum-0.9 - acc/train"]
```

```python
epochs = np.array(list(np.arange(0,200,10)) + [199])
print(epochs)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.grid(True)
fontsize = 30
markersize = 15
markevery = 5

plt.plot(epochs, resnet18_learnable_train[epochs]*100,  linewidth=5, alpha=1, marker = "o", markersize = markersize, markevery = markevery)
plt.plot(epochs, CMF_resnet18_m1_train[epochs]*100, linewidth=5, alpha=1, marker = "X", markersize = markersize, markevery = markevery)
plt.plot(epochs, CMF_resnet18_m9_train[epochs]*100, linewidth=5, alpha=1, marker = "P", markersize = markersize, markevery = markevery)

plt.xlabel('Epochs', fontsize=fontsize)
plt.ylabel(r'Train Acc', fontsize=fontsize)

plt.xticks(np.arange(0, 250, 50), fontsize=fontsize)
plt.yticks(np.arange(0, 110, 20), fontsize=fontsize)

plt.legend(['Learnable Classifier', r'CMF, $\beta$ = 0.1', r'CMF, $\beta$ = 0.9'], fontsize=fontsize, loc='lower right')

#plt.savefig(f"{save_path}/CMF_resnet18_train.png", dpi = 600, bbox_inches='tight')
plt.show()
```
