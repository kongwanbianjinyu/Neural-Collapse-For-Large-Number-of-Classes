```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
import os
save_path = "./images/learning_curve"
#os.mkdir(save_path)
```

```python
pd.read_csv("training_curve.csv")
```

```python
df = pd.read_csv("training_curve.csv")
```

```python
print(df.keys())
```

```python
resnet18_K100_train = df["resnet18-cifar100-d10-tau10.0-epoch200-lr0.1 - acc/train"]
densenet121_K100_train = df["densenet121-cifar100-d10-tau10.0-epoch200-lr0.1 - acc/train"]
resnext50_K100_train = df["resnext50-cifar100-d10-tau10.0-epoch200-lr0.1 - acc/train"]
```

```python
resnet18_K200_train = df["resnet18-tiny_imagenet-d10-tau10.0-epoch200-lr0.1-val - acc/train"]
densenet121_K200_train = df["densenet121-tiny_imagenet-d10-tau10.0-epoch200-lr0.1-val - acc/train"]
resnext50_K200_train = df["resnext50-tiny_imagenet-d10-tau10.0-epoch200-lr0.1-val - acc/train"]
```

```python
epochs = np.array(list(np.arange(0,200,10)) + [199])
print(epochs)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.grid(True)
fontsize = 30
markersize = 15
markevery = 5

plt.plot(epochs, resnet18_K100_train[epochs]*100,  linewidth=5, alpha=1, marker = "o", markersize = markersize, markevery = markevery)
plt.plot(epochs, densenet121_K100_train[epochs]*100, linewidth=5, alpha=1, marker = "X", markersize = markersize, markevery = markevery)
plt.plot(epochs, resnext50_K100_train[epochs] *100, linewidth=5, alpha=1, marker = "P", markersize = markersize, markevery = markevery)

plt.xlabel('Epochs', fontsize=fontsize)
plt.ylabel(r'Train Acc', fontsize=fontsize)

plt.xticks(np.arange(0, 250, 50), fontsize=fontsize)
plt.yticks(np.arange(0, 110, 20), fontsize=fontsize)

plt.legend(['ResNet18', r'DenseNet121', r'ResNeXt50'], fontsize=fontsize, loc='lower right')

plt.savefig(f"{save_path}/CIFAR100_train.png", dpi = 600, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.grid(True)
fontsize = 30
markersize = 15
markevery = 5

plt.plot(epochs, resnet18_K200_train[epochs]*100,  linewidth=5, alpha=1, marker = "o", markersize = markersize, markevery = markevery )
plt.plot(epochs, densenet121_K200_train[epochs]*100, linewidth=5, alpha=1, marker = "X", markersize = markersize, markevery = markevery )
plt.plot(epochs, resnext50_K200_train[epochs] *100, linewidth=5, alpha=1, marker = "P", markersize = markersize, markevery = markevery )

plt.xlabel('Epochs', fontsize=fontsize)
plt.ylabel(r'Train Acc', fontsize=fontsize)

plt.xticks(np.arange(0, 250, 50), fontsize=fontsize)
plt.yticks(np.arange(0, 110, 20), fontsize=fontsize)

plt.legend(['ResNet18', r'DenseNet121', r'ResNeXt50'], fontsize=fontsize, loc='lower right')

plt.savefig(f"{save_path}/TinyImageNet_train.png", dpi = 600, bbox_inches='tight')
plt.show()
```
